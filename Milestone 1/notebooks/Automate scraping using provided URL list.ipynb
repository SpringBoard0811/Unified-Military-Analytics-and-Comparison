# This function fetches a webpage from the given URL and saves its HTML content to a file, handling errors and failed requests.

```python
def download_page(url, filename):
    try:
        res = requests.get(url, timeout=15)
        if res.status_code == 200:
            with open(filename, "w", encoding="utf-8") as f:
                f.write(res.text)
        else:
            print(f"Failed: {url}")
    except Exception as e:
        print(f"Error {url} -> {e}")
```

```python
download_page(base_url, "gfp_pages/countries_listing.html")
```

# Returns None for None input; otherwise strips commas/spaces, extracts the first number and unit with regex, and joins them into a single string.

```python
# Unit Parser
def unit(raw_text):
    if raw_text is None:
        return None

    text = raw_text.replace(",", "").strip()

    number_match = re.search(r"[\d.]+", text)
    number = number_match.group() if number_match else ""

    unit_match = re.search(r"[a-zA-Z%./]+", text)
    unit = unit_match.group() if unit_match else ""

    return f"{number} {unit}".strip()
```

# Reads an HTML file, finds all metric cards, extracts each card’s country and metric value using the unit parser, and returns a DataFrame with countries and the specified metric.

```python
def parse_metric_cards(filepath, metric_name):
    with open(filepath, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "lxml")

    cards = soup.find_all("div", class_="recordsetContainer")
    data = []

    for card in cards:
        # country name
        country_tag = card.find("div", class_="longFormName")
        if not country_tag:
            continue
        country = country_tag.get_text(strip=True)

        # metrics values
        value_tag = card.find("div", class_="valueContainer")
        if not value_tag:
            continue

        # units
        raw_text = value_tag.get_text(" ", strip=True)
        value = unit(raw_text)

        data.append([country, value])

    df = pd.DataFrame(data, columns=["country", metric_name])
    return df
```

#Iterates over all HTML files in the metrics folder, parses each into a DataFrame using its filename as the metric name, and merges them all on `country` to build a combined `master_df`.


```python
METRIC_FOLDER = "gfp_pages/metrics"
files = [f for f in os.listdir(METRIC_FOLDER) if f.endswith(".html")]

master_df = None

for file in tqdm(files):
    metric_name = file.replace(".html", "")
    filepath = os.path.join(METRIC_FOLDER, file)

    df_metric = parse_metric_cards(filepath, metric_name)

    if master_df is None:
        master_df = df_metric
    else:
        master_df = pd.merge(master_df, df_metric, on="country", how="outer")
```

# Looks up the row for the given country in df, and if found returns it transposed with the country as the column name; otherwise returns "Country not found"

```python
# metrics for any country
def get_country_metrics(country_name):
    row = df[df["country"] == country_name]
    if row.empty:
        return "Country not found"
    result = row.T
    result.columns = [country_name]
    return result
```

# Returns 0 for NaN input; otherwise extracts the first number from the text and converts it to an integer, or 0 if no number is found.

```python
def extract_number(text):
    if pd.isna(text):
        return 0
    match = re.search(r"\d+", str(text))
    return int(match.group()) if match else 0
```

# Applies extract_number to the aircraft_carriers column and stores the resulting numbers in a new column aircraft_carriers_num

```python
df["aircraft_carriers_num"] = df["aircraft_carriers"].apply(extract_number)
```

# Sorts the DataFrame by aircraft_carriers_num descending, selects only country and aircraft_carriers_num columns, and returns the top 10 rows.

```python
top10 = (
    df.sort_values(by="aircraft_carriers_num", ascending=False)
      [["country", "aircraft_carriers_num"]]
      .head(10)
)

top10
```
