# KPI Generation and Composite Military Strength Index

This notebook generates normalized performance indicators (KPIs)
to remove scale bias and enable fair comparison between countries.
These KPIs are later combined into a composite military strength score.

### Import required libraries

```python
import pandas as pd
import numpy as np


```

### Reading the Cleaned Dataset

```python
df = pd.read_csv("military_cleaned.csv")
df.columns
```

### KPI 1: Power Index Rank Gap

Formula:
Power Index Rank - 1

Why this KPI?

The Power Index Rank indicates a country’s global military position.
By subtracting 1, we measure how far a country is from the top-ranked nation.

A value of 0 means the country is ranked #1.
Higher values indicate increasing distance from the global leader.

This converts ordinal ranking into a measurable gap indicator,
which is easier to compare and visualize in dashboards.

```python
# Power Index Rank Gap
df["power_index_rank_gap"] = df["power_index_rank"] - 1
df
```

### KPI 2: Assets per Capita

Formula:
Total Military Assets / Total Population

Why this KPI?

Absolute asset count favors large countries.
This KPI measures military asset availability relative to population size,
showing military intensity instead of raw volume.

It helps identify smaller nations with disproportionately strong military capacity.

```python
df["assets_per_capita"] = df["total_assets"] / df["total_population"]
df
```

### KPI 3: Defense Budget to Purchasing Power Ratio

Formula:
Defense Budget / Purchasing Power Parity (PPP)

Why this KPI?

Raw defense budget alone is misleading because
countries have different economic strengths.

This KPI shows how much economic power is allocated
toward military expenditure, reflecting defense priority intensity.

```python
df["budget_to_gdp_ratio"] = (
    df["defense_budget"] / df["purchasing_power_parity"]
)
df
```

### KPI 4: Active Personnel Ratio

Formula:
Active Personnel / Total Population

Why this KPI?

Large countries naturally have more soldiers.
This KPI measures military mobilization level relative to population,
indicating how militarized a nation is proportionally.

```python
df["active_personnel_ratio"] = (
    df["active_personnel"] / df["total_population"]
)
df
```

### KPI 5: Available Manpower Ratio

Formula:
Available Manpower / Total Population

Why this KPI?

This measures the reserve capacity of a country.
It reflects long-term war sustainability and mobilization potential.

```python
df["manpower_ratio"] = (
    df["available_manpower"] / df["total_population"]
)
df
```

```python
#Additional

df["defense_to_gdp_ratio"] = df["defense_budget"] / df["purchasing_power_parity"]

df["oil_self_sufficiency"] = df["oil_production"] / df["oil_consumption"]

```

```python
# Rounding KPI Columns
kpi=["power_index_rank_gap",
    "assets_per_capita",
    "budget_to_gdp_ratio","active_personnel_ratio",
    "manpower_ratio","defense_to_gdp_ratio",
    "oil_self_sufficiency"
]
   
df[kpi]=df[kpi].round(4)
```

```python
print(df[[
    "power_index_rank_gap",
    "assets_per_capita",
    "budget_to_gdp_ratio"
]].head())

```

### Metadata Enrichment

In this step, additional contextual information is added to the dataset 
without altering the core numerical military metrics.

Metadata enrichment includes attributes such as:
- Geographic classification (e.g., continent or region)
- Strategic alliances (e.g., NATO membership)

Why this step is important?

Raw military numbers lack context. 
Two countries may have similar military assets but operate in completely different geopolitical environments.

Metadata allows:
- Group-wise comparisons (region-wise, alliance-wise analysis)
- Segmented visualization
- Context-aware KPI evaluation
- Better strategic interpretation

Important:

Metadata does not change numerical values.
It enhances interpretability and analytical depth.
