# Military Data Cleaning

This notebook processes the raw scraped dataset.
The workflow includes data cleaning, column filtering, renaming the columns and sorting them alphabetically.

### Import Required Libraries

```python
import pandas as pd
import re
```

### Load Raw Scraped Dataset

The raw dataset generated from the scraping pipeline
is loaded into a DataFrame for cleaning and processing.

```python
df = pd.read_csv("military_raw_data.csv")

df.head()

```

```python
#standardize everything
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_", regex=False)
    .str.replace("-", "_", regex=False)
    .str.replace("/", "_", regex=False)
    .str.replace(".", "", regex=False)
    .str.replace("*", "", regex=False)
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)
df.head()
```

### Initial Data Cleaning

We:
- Remove commas from numeric values
- Convert relevant columns to numeric format
- Handle missing values

```python
#standardize everything
def clean_numeric(value):
    if pd.isna(value):
        return None
    value = re.sub(r"[^\d.]", "", str(value))
    return float(value) if value else None

for col in df.columns:
    if col != "country":
        df[col] = df[col].apply(clean_numeric)

df.head()
```

### Remove the "2026" & "Military Strength" from "Country"

```python
df = (
    df.assign(
        country=df["country"]
        .str.replace(r"\s*\d{4}", "", regex=True)
        .str.strip()
    ))
df = (
    df.assign(
        country=df["country"]
        .str.replace(r"Military Strength", "", case=False)
        .str.replace(r"\s*\d{4}", "", regex=True)
        .str.strip()
    )    
    .sort_values("country")
    .reset_index(drop=True))

df.head()
```

```python
# Finalized column list
final_columns = [
    "country",
    "power_index_score",
    "power_index_rank",

    "active_personnel",
    "reserve_personnel",
    "available_manpower",
    "total_population",

    "aircraft_total",
    "fighters",
    "attack_helicopters",

    "tanks",
    "self_propelled_artillery",
    "mlrs_rocket_artillery",

    "total_assets",
    "submarines",
    "destroyers",
    "frigates",
    "aircraft_carriers",

    "defense_budget",
    "purchasing_power_parity",
    "foreign_exchange_gold",

    "oil_production",
    "oil_consumption",
    "oil_proven_reserves",
    "natural_gas_production",
    "natural_gas_consumption"
]

# Keep only finalized columns
df = df[final_columns]

# Reset index for clean structure
df = df.reset_index(drop=True)

# Quick verification
df.info()
df.head()
```

### Handle Missing Values
#Drop columns with>50% missing values

#We replace missing numeric values (`NaN`) with the median of each column.  
The median is chosen because it is robust to extreme values and reduces distortion caused by outliers.  
This ensures the dataset contains no missing values before further analysis or modeling.  

```python
missing_percent = df.isna().mean() * 100
cols_to_drop = missing_percent[missing_percent > 50].index
df.drop(columns=cols_to_drop, inplace=True)
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

### Data Ordering and Column Alignment

```python

df = df.sort_values("country")
df = df.reset_index(drop=True)
df = df.reindex(sorted(df.columns), axis=1)
df = df[["country"] + [col for col in df.columns if col != "country"]]


df.head()
```

### Export Cleaned Dataset

The final processed dataset is exported
for Tableau dashboard and visualization usage.

```python
df.to_csv("military_cleaned.csv", index=False)
```
