## KPI Engineering and Power BI Preparation

### KPI Feature Engineering

This represents the **KPI Engineering stage** of the project. In this phase, the cleaned military dataset obtained from the previous milestone1 is processed further to generate **analytical metrics and key performance indicators (KPIs).**

The main goal of this step is to transform the prepared dataset into meaningful indicators that can be used for **interactive visualization and analysis in a Power BI dashboard.**

In this notebook, the following operations are performed:

- Import the final cleaned military dataset  
- Calculate important global KPIs such as:
  - The country with the highest military ranking
  - Total count of countries in the dataset
  - Overall global defense budget
  - Total global military manpower
- Generate additional derived analytical attributes
- Export the updated dataset containing KPI values for dashboard development

These computed KPIs provide valuable insights into **how military strength, defense spending, and manpower are distributed among different countries worldwide.**

## Step 2 – Define Dataset Path

In this step, the file path for the **final cleaned military dataset** created during the earlier data preprocessing phase is specified.

- Military power rankings  
- Defense spending or budgets  
- Number of active military personnel  
- Additional defense-related statistics

```python
DATA_PATH = "Final_Military_Data.csv"
```

```python
df = pd.read_csv(DATA_PATH)

print("Dataset loaded successfully.")
print(f"Total records: {len(df)}")
```

# Step 3 – KPI Calculation

In this step, multiple **Key Performance Indicators (KPIs)** are calculated to summarize important global military statistics.

The purpose of generating these KPIs is to convert raw numerical values into **meaningful analytical insights** that can support further analysis and visualization.

### KPIs Generated

**1️⃣ Top Country**

Determines the country that holds the **highest military power position**, which corresponds to the lowest ranking value in the dataset.

**2️⃣ Total Countries**

Counts the total number of unique countries present in the dataset.

**3️⃣ Selected Rank**

Adds a derived field that represents the military power ranking assigned to each country.

**4️⃣ Global Defense Budget**

Calculates the **combined defense spending of all countries included in the dataset**.

**5️⃣ Global Military Manpower**

Computes the total number of **active military personnel across all countries globally**.

These KPIs offer a clear summary of **worldwide military strength, defense spending, and manpower distribution**.

```python
# KPI 1: Top Country

top_country = df.loc[df['p_rank'].idxmin(), 'country']

# KPI 2: Total Countries
total_countries = df['country'].nunique()

# KPI 3: Selected Rank (Country Rank)
df['selected_rank'] = df['p_rank']

# KPI 4: Global Defense Budget
global_budget = df['defense_budget_usd'].sum()

# KPI 5: Global Manpower
global_manpower = df['active_personnel'].sum()

# Display KPIs
print("\n===== GLOBAL KPIs =====")

print("Top Country:", top_country)
print("Total Countries:", total_countries)
print("Global Defense Budget:", global_budget)
print("Global Manpower:", global_manpower)

# Save Dataset with KPIs

df.to_csv("military_kpi_dataset.csv", index=False)

print("\nKPI Dataset Saved: military_kpi_dataset.csv")
```
