## Module 3: KPI Feature Engineering 

This notebook implements the **KPI Engineering phase** of the project. In this milestone, the cleaned military dataset generated in the previous stage is transformed into **analytical metrics and key performance indicators (KPIs)**.

The objective of this notebook is to compute **global defense statistics and analytical indicators** that will later be used in the **Power BI dashboard for interactive military analytics**.

The program performs the following tasks:

- Load the final cleaned military dataset
- Compute global-level KPIs such as:
  - Top ranked military country
  - Total number of countries
  - Global defense budget
  - Global military manpower
- Create derived analytical fields
- Export a KPI-enriched dataset for dashboard visualization

These KPIs help provide insights into **global military power distribution, defense investment, and manpower capacity across countries.**


## Step 1 – Import Required Libraries

In this step, the required Python libraries for data processing and analysis are imported.

```python
import pandas as pd
```

## Step 2 – Define Dataset Path

This step defines the location of the **final cleaned military dataset** generated during the data preprocessing stage.

The dataset contains structured military statistics for multiple countries, including metrics such as:

- Military power ranking
- Defense budgets
- Active military personnel
- Other defense-related indicators

The dataset will be used as the **input source for generating analytical KPIs**.


```python
DATA_PATH = "Final_Military_Data.csv"
```

```python
df = pd.read_csv(DATA_PATH)

print("Dataset loaded successfully.")
print(f"Total records: {len(df)}")
```

## Step 3 – KPI Calculation

This section computes several **Key Performance Indicators (KPIs)** that summarize global military statistics.

These KPIs help transform raw numerical data into **actionable analytical insights**.

### KPIs Generated

**1️⃣ Top Country**

Identifies the country with the **highest military power ranking** (lowest rank value).

**2️⃣ Total Countries**

Calculates the number of distinct countries available in the dataset.

**3️⃣ Selected Rank**

Creates a derived column representing each country's military power rank.

**4️⃣ Global Defense Budget**

Computes the **total defense expenditure across all countries**.

**5️⃣ Global Military Manpower**

Calculates the total number of **active military personnel worldwide**.

These indicators provide an overview of **global defense capabilities and investment levels**.

```python
# KPI 1: Top Country

top_country = df.loc[df['p_rank'].idxmin(), 'country']

# KPI 2: Total Countries
total_countries = df['country'].nunique()

# KPI 3: Selected Rank (Country Rank)
df['selected_rank'] = df['p_rank']

# KPI 4: Global Defense Budget
global_budget = df['defense_budget_usd'].sum()

# KPI 5: Global Manpower
global_manpower = df['active_personnel'].sum()

# Display KPIs
print("\n===== GLOBAL KPIs =====")

print("Top Country:", top_country)
print("Total Countries:", total_countries)
print("Global Defense Budget:", global_budget)
print("Global Manpower:", global_manpower)

# Save Dataset with KPIs

df.to_csv("military_kpi_dataset.csv", index=False)

print("\nKPI Dataset Saved: military_kpi_dataset.csv")
```
