# 🌍 Step 1 – Load Raw Military Data
**Military Dataset Inspection**

This step loads the raw military dataset and inspects its structure, column names, data types, and sample rows.  
*Purpose:* Identify initial data quality and columns needing cleaning.

```python

import pandas as pd

df = pd.read_csv("Military_raw_data (6).csv")
print(df.head())
print(df.info())
```

# 🔍 Step 2 – Identify Data Noise
**Check for Inconsistencies**

Review all column names and detect noisy numeric fields (text, commas, units, symbols).  
*Purpose:* Ensure numeric columns are ready for processing.

```python

print(df.columns.tolist())
print(df.info())
```

# 🧹 Step 3 – Clean Numeric Columns
**Transform Text to Numbers**

Remove non-numeric characters and convert columns to float.  
*Purpose:* Enable calculations and KPI engineering without errors.

```python

def clean_numeric(col):
    return (
        col.astype(str)
           .str.replace(r'[^0-9.]', '', regex=True)
           .replace('', '0')
           .astype(float)
    )

for col in df.columns:
    if col != 'country':  # Exclude country names
        df[col] = clean_numeric(df[col])
```

# ✨ Step 4 – Standardize Column Names
**Consistency Across Columns**

Strip extra spaces and convert all column names to lowercase.  
*Purpose:* Avoid issues during coding and analysis.

```python

df.columns = df.columns.str.strip().str.lower()
```

# 💧 Step 5 – Handle Missing Values
**Fill Gaps**

Replace missing values with zero.  
*Purpose:* Prevent calculation errors in subsequent KPI computations.

```python

df = df.fillna(0)
```

# ✅ Step 6 – Validate Cleaned Data
**Data Integrity Checks**

Check data types, missing values, duplicates, and basic statistics.  
*Purpose:* Ensure dataset is clean and reliable for analysis.

```python

df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

# 💾 Step 7 – Export Cleaned Dataset
**Save for Dashboard Integration**

Export cleaned dataset to a CSV file.  
*Purpose:* Ready the dataset for KPI engineering and reporting.

```python

df.to_csv("military_data_cleaned.csv", index=False)
```

##Module-3 KPI Engineering & Feature Creation


# 📊 Step 8 – Load Clean Dataset
**Ready for KPI Engineering**

Load the cleaned dataset to compute derived metrics and KPIs.  
*Purpose:* Prepare data for analysis and visualization.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("military_data_cleaned.csv")
print(df.head())
```

# ⚡ Step 9a – Assets per Capita
**Measure Equipment Strength**

Compute total military assets per soldier.  
*Purpose:* Compare equipment strength across countries.

```python

import numpy as np

df["assets_per_capita"] = (
    df["aircraft_total"] +
    df["tanks"] +
    df["submarines"]
) / df["available_manpower"].replace(0, np.nan)
```

# 💰 Step 9b – Budget-to-GDP Ratio
**Economic Investment in Defense**

Calculate defense spending relative to national GDP (PPP-adjusted).  
*Purpose:* Assess defense investment intensity.

```python

df["budget_to_gdp_ratio"] = (
    df["defense_budget"] /
    df["purchasing_power_parity"].replace(0, np.nan)
)
```

# 🏆 Step 9c – Power Index Rank Gap
**Compare Relative Ranking**

Rank countries by defense budget and compute the gap from the top-ranked nation.  
*Purpose:* Facilitate comparative analysis.

```python

df["rank"] = df["defense_budget"].rank(ascending=False)
top_rank = df["rank"].min()
df["power_index_rank_gap"] = df["rank"] - top_rank
```

# 🛡 Step 10 – Strength Category
**Classify Military Capability**

Categorize countries into Low, Medium, High tiers using assets per capita.  
*Purpose:* Provide a simple visual indicator of military strength.

```python

df["strength_category"] = pd.cut(
    df["assets_per_capita"],
    bins=3,
    labels=["Low", "Medium", "High"]
)
```

# 🔍 Step 11 – Validate KPI Calculations
**Check for Errors**

Inspect for infinite or unrealistic values, and verify a sample of countries.  
*Purpose:* Ensure KPIs are logical and accurate.

```python

print("Infinite values check:\n")
print(np.isinf(df.select_dtypes(include=[float, int])).sum())

print("\nKPI Statistics:\n")
print(df[["assets_per_capita", "budget_to_gdp_ratio", "power_index_rank_gap"]].describe())

print("\nSample Data:\n")
print(df[["country", "assets_per_capita", "budget_to_gdp_ratio", "power_index_rank_gap"]].head(5))
```

**Final Dataset for Analysis**

Save the dataset with all engineered KPIs for dashboards, reporting, and visualization.  
*Purpose:* Make the dataset ready for integration into analytics workflows.
