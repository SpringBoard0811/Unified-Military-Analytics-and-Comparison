### Data Cleaning & Standardization

Load Raw Data & Basic Analysis

This step loads the raw scraped military dataset

```python
import pandas as pd
df = pd.read_csv("military data.csv")
```

The dataset was successfully loaded and inspected.

Performs an initial inspection to understand its structure, column names, and data types.

```python
print(df.head())
print(df.info())
```

Several numeric columns were stored as object type and required cleaning.

Identify Data Noise: This step reviews column names and data types to identify noisy fields and incorrect formats.

```python
df.columns.tolist()
df.info()
```

Column inspection confirmed the presence of text-formatted numeric values and inconsistent column formatting.

Clean Numeric Columns: This step removes non-numeric characters such as symbols, commas, and units from numeric columns and converts them into float format.

```python
def clean_numeric(col):
    return (
        col.astype(str)
           .str.replace(r'[^0-9.]', '', regex=True)
           .replace('', '0')
           .astype(float)
    )

for col in df.columns:
    if col != 'country':
        df[col] = clean_numeric(df[col])
```

All numeric columns were cleaned and converted into proper numerical format, making them suitable for analysis.

Standardize Column Names: This step standardizes column names by removing extra spaces and converting them to lowercase for consistency.

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
)
```

Column names were standardized to maintain consistency across the dataset.

Handle Missing Values: This step handles missing values to prevent errors during KPI calculations.

```python
df = df.fillna(0)
```

Missing values were replaced with zero to maintain dataset completeness.

Validate Cleaned Data: This step validates the cleaned dataset by checking data types, missing values, duplicates, and statistical summaries.

```python
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

The dataset was validated successfully, ensuring no missing values, no duplicates, and correct data types.

Export Cleaned Dataset: This step exports the cleaned dataset for further KPI engineering.

```python
df.to_csv("military_data_cleaned.csv", index=False)
```

The cleaned dataset was exported successfully for advanced analysis.


### KPI Engineering & Feature Creation

Load Clean Dataset: This step loads the cleaned dataset to begin KPI engineering.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("military_data_cleaned.csv")
print(df.head())
```

The cleaned dataset was loaded successfully and is ready for KPI creation.

Define Required KPIs

Assets per Capita   (total_aircraft + tanks + naval_assets) / total_manpower.   This step calculates Assets per Capita to measure total military assets relative to total manpower.

```python
df["assets_per_capita"] = (
    df["total_military_aircraft"] +
    df["tanks"] +
    df["total_naval_fleet"]
) / df["total_military_manpower"].replace(0, np.nan)
```

Assets per Capita was successfully computed as an indicator of equipment strength.

Budget-to-GDP Ratio   defense_budget / gdp. This step computes Budget-to-GDP Ratio to evaluate defense spending relative to economic strength.

```python
df["budget_to_gdp_ratio"] = (
    df["defense_budget_usd"] /
    df["purchasing_power_parity_usd"].replace(0, np.nan)
)
```

Budget-to-GDP Ratio was derived to assess defense investment intensity.

Power Index Rank Gaprank(country) – rank(top_country). This step ranks countries based on defense budget and calculates the rank gap compared to the top country.

```python
df["rank"] = df["defense_budget_usd"].rank(ascending=False)

top_rank = df["rank"].min()

df["power_index_rank_gap"] = df["rank"] - top_rank
```

Ranking and rank gap were successfully computed to support comparative analysis.

Metadata Enrichment

Strength Category Based on Assets per Capita ,This step categorizes countries into strength tiers based on Assets per Capita.

```python
df["strength_category"] = pd.cut(
    df["assets_per_capita"],
    bins=3,
    labels=["Low", "Medium", "High"]
)
```

Countries were categorized into Low, Medium, and High strength tiers.

Validate KPI Values

Check division by zero, This step validates the KPI calculations by checking for division-by-zero errors. During KPI computation, division operations may produce infinite values if the denominator is zero. Therefore, we inspect all numeric columns to ensure no infinite values exist.

```python
import numpy as np

print("Infinite values check:\n")
print(np.isinf(df.select_dtypes(include=[float, int])).sum())
```

The output confirms whether any infinite values are present in the dataset. If all values are zero, it indicates that no division-by-zero errors occurred and the KPI calculations are numerically valid.
