## Data Cleaning & Standardization
#### Load Raw Data & Basic Analysis

```python
import pandas as pd
import numpy as np

df = pd.read_excel("military_raw_collected.xlsx")
```

```python
print(df.head())
print(df.info())
```

Identify Data Noise: This step reviews column names and data types to identify noisy fields and incorrect formats.

```python
df.columns.tolist()
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

##### This step removes non-numeric characters such as symbols, commas, and units from numeric columns and converts them into float format.



```python
def clean_numeric(col):
   return(
       col.astype(str)
          .str.replace(r'[^0-9.]', '', regex=True)
          .replace('','0')
          .astype(float)
   )
for col in df.columns:
    if col != 'country':
        df[col]=clean_numeric(df[col])
```

######  This step standardizes column names by removing extra spaces and converting them to lowercase for consistency.

```python
df.columns=(
    df.columns
      .str.strip()
      .str.lower()
)
```

```python
df=df.fillna(0)
```

#### This step validates the cleaned dataset by checking data types, missing values, duplicates, and statistical summaries.

```python
df.info()
df.describe()
df.isnull().sum
df.duplicated().sum()
```

#### This step exports the cleaned dataset for further KPI engineering.

```python
df.to_csv("military_data_cleaned.csv",index=False)
```

#### KPI Engineering & Feature Creation

This step loads the cleaned dataset to begin KPI engineering

```python
import pandas as pd
import numpy as np
df=pd.read_csv("military_data_cleaned.csv")
print(df.head())
```

This step calculates Assets per Capita to measure total military assets relative to total manpower.

```python
df["assets_per_capita"]=(
    df["total_military_aircraft"]+
    df["tanks"]+
    df["total_naval_fleet"]
) / df["total_military_manpower"].replace(0,np.nan)
```

This step computes Budget-to-GDP Ratio to evaluate defense spending relative to economic strength.

```python
df["budget_to_gdp_ratio"]=(
    df["defense_budget_usd"] /
    df["purchasing_power_parity_usd"].replace(0,np.nan)
)
```

This step ranks countries based on defense budget and calculates the rank gap compared to the top country.

```python
df["rank"]=df["defense_budget_usd"].rank(ascending=False)
top_rank=df["rank"].min()
df["power_index_rank_gdp"]=df["rank"] - top_rank
```

This step categorizes countries into strength tiers based on Assets per Capita.

```python
df["strength_category"]=pd.cut(
    df["assets_per_capita"],
    bins=3,
    labels=["Low","Medium","High"]
)
```
