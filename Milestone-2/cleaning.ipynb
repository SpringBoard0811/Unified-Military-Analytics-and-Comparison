Step 1: Load Raw Data & Basic Analysis

```python
import pandas as pd

df = pd.read_csv("military data.csv")
print(df.head())
print(df.info())
```

Identify Data Noise

```python
df.columns.tolist()
df.info()
```

Step 3: Clean Numeric Columns

```python
def clean_numeric(col):
    return (
        col.astype(str)
           .str.replace(r'[^0-9.]', '', regex=True)
           .replace('', '0')
           .astype(float)
    )

for col in df.columns:
    if col != 'country':
        df[col] = clean_numeric(df[col])
```

Step 4: Standardize Column Names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
)
```

Step 5: Handle Missing Values

```python
df = df.fillna(0)
```

Step 6: Validate Cleaned Data

```python
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

Step 7: Export Cleaned Dataset

```python
df.to_csv("military_data_cleaned.csv", index=False)
```

STEP 1: Load Clean Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("military_data_cleaned.csv")

print(df.head())
```

STEP 2: Define Required KPIs

Assets per Capita   (total_aircraft + tanks + naval_assets) / total_manpower

```python
df["assets_per_capita"] = (
    df["total_military_aircraft"] +
    df["tanks"] +
    df["total_naval_fleet"]
) / df["total_military_manpower"].replace(0, np.nan)
```

Budget-to-GDP Ratio   defense_budget / gdp

```python
df["budget_to_gdp_ratio"] = (
    df["defense_budget_usd"] /
    df["purchasing_power_parity_usd"].replace(0, np.nan)
)
```

Power Index Rank Gaprank(country) – rank(top_country)

```python
df["rank"] = df["defense_budget_usd"].rank(ascending=False)

top_rank = df["rank"].min()

df["power_index_rank_gap"] = df["rank"] - top_rank
```
