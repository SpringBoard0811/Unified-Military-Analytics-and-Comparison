````python
import requests
from bs4 import BeautifulSoup

url="https://www.globalfirepower.com/country-military-strength-detail.php?country_id=russia"

page = requests.get(url)

soup = BeautifulSoup(page.text, "html")
soup
````

````python
c_span = soup.find('h1')

country_text = c_span.get_text(strip=True)

country = country_text.replace("2026", "").replace("Military Strength", "").strip()

print(country)
````

````python
main_span = soup.find('span', class_='textNormal textDkGray')

f = main_span.find_all('span', class_='textBold')
#print(f)
top_rank_text = f[0].get_text(strip=True)
top_rank = top_rank_text.replace("Top", "").strip()
print(top_rank)
````

````python
country_rank = f[2].get_text(strip=True)
print(country_rank)
````

````python
power_index_rank = f[3].get_text(strip=True)
print(power_index_rank)
````

````python
#defense_budget
defence_span = soup.find_all('span', class_='textWhite textShadow')
# print(defence_span[2])
defence_budget = defence_span[2].text
print(defence_budget)
# print(defence_span)
````

````python
#population
population_span = soup.find_all('span', class_='textLarge textWhite textShadow')
# print(population_span)

p_text=population_span[0].text
population = p_text.replace("Population: ", "").strip()
print(population)
````

````python
#available_manpower, active_personnel, reserve_personnel
manpower_span = soup.find_all('span', class_='textWhite textShadow')
# print(manpower_span)
manpower = manpower_span[8].text.strip()
active_personnels = manpower_span[12].text.strip()
reserve_personnel = manpower_span[13].text.strip()
paramilitory_personnel = manpower_span[14].text.strip()
airforce_personnel = manpower_span[15].text.strip()
army_personnel = manpower_span[16].text.strip()
navy_personnel = manpower_span[17].text.strip()

print(manpower)
print(active_personnels)
print(reserve_personnel)
print(paramilitory_personnel)
print(airforce_personnel)
print(army_personnel)
print(navy_personnel)
````

````python
#total aircraft
tair_span = soup.find_all('span', class_='textLarge textWhite textShadow')
# print(tair_span)

total_aircraft = tair_span[1].text.split("\n")[0]
total_aircraft = total_aircraft.replace("Stock: ", "").strip()
print(total_aircraft)
````

````python
#fighter_aircraft, tank
air_span = soup.find_all('span', class_='textWhite textShadow')
# print(air_span)
# print(air_span[27])

text = air_span[19].text

fighter_aircraft = text.split("\n")[0]          # take first line
fighter_aircraft = fighter_aircraft.split("(")[0]    # remove percentage
fighter_aircraft = fighter_aircraft.replace("Stock:", "").strip()
print(fighter_aircraft)

text = air_span[27].text
tank = text.split("\n")[0]          # take first line
tank = tank.split("(")[0]    # remove percentage
tank = tank.replace("Stock:", "").strip()
print(tank)


````

````python
#naval_assets
naval_span = soup.find_all('span', class_='textLarge textWhite textShadow')
naval_assets = naval_span[2].text
print(naval_assets)
````

````python
print("="*60)
print(f"        🇮🇳 {country} Military Overview")
print("="*60)

print("\n📊 Global Ranking")
print("-"*60)
print(f"Top Rank Category   : {top_rank}")
print(f"Country Rank        : {country_rank}")
print(f"Power Index Score   : {power_index_rank}")

print("\n💰 Defence & Population")
print("-"*60)
print(f"Defence Budget      : {defence_budget}")
print(f"Population          : {population}")

print("\n👥 Manpower")
print("-"*60)
print(f"Total Manpower      : {manpower}")
print(f"Active Personnel    : {active_personnels}")
print(f"Reserve Personnel   : {reserve_personnel}")
print(f"Paramilitary        : {paramilitory_personnel}")
print(f"Airforce Personnel  : {airforce_personnel}")
print(f"Army Personnel      : {army_personnel}")
print(f"Navy Personnel      : {navy_personnel}")

print("\n✈️  Air & Land Strength")
print("-"*60)
print(f"Total Aircraft      : {total_aircraft}")
print(f"Fighter Aircraft    : {fighter_aircraft}")
print(f"Tanks               : {tank}")

print("\n🚢 Naval Strength")
print("-"*60)
print(f"Total Naval Assets  : {naval_assets}")

print("\n" + "="*60)
````
