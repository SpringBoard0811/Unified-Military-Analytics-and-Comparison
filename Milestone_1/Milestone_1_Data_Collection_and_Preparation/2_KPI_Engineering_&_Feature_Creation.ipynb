```python
import pandas as pd

df = pd.read_csv('/content/Military_clean_data.csv')
print("Data loaded successfully into a DataFrame named 'df'.")
df
```

```python
# normalized assets by manpower (per capita approach)

# calculated defense spending intensity using GDP ratio

# derived relative competitiveness using rank gap

# 1. Assets per capita
df['assets_per_capita'] = (
    df['total_military_aircraft'] +
    df['tanks'] +
    df['total_naval_fleet']
) / df['total_military_manpower']

# 2. Budget to GDP ratio
df['budget_to_gdp_ratio'] = (
    df['defense_budget_usd'] /
    df['purchasing_power_parity_usd']
)

# 3. Power score & rank
df['power_score'] = (
    df['total_military_aircraft'] +
    df['tanks'] +
    df['total_naval_fleet']
)

df['rank'] = df['power_score'].rank(ascending=False)
df['rank_gap'] = df['rank'] - df['rank'].min()

df[['Country_Code','assets_per_capita','budget_to_gdp_ratio','power_score']]
```
