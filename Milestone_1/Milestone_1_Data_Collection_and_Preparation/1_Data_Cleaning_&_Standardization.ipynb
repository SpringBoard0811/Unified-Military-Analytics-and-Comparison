```python
df['Country_Code']
df['Country_Name']
df['total_population']

import pandas as pd

def clean_numeric_column(df, column_name):
    df[column_name] = pd.to_numeric(
        df[column_name]
        .astype(str)                      # ensure string
        .str.replace(r'[^\d.]', '', regex=True),  # remove non-numeric chars
        errors='coerce'
    )
    return df

def clean_text_column(df, column_name):
    df[column_name] = pd.to_numeric(
        df[column_name]
        .astype(str)                              # ensure string
        .str.replace(r'[^\d.]', '', regex=True),  # keep only digits & decimal
        errors='coerce'
    )
    return df

df = clean_numeric_column(df, 'total_population')
df = clean_numeric_column(df, 'total_military_manpower')
df = clean_numeric_column(df, 'population_reaching_military_age_annually')
df = clean_numeric_column(df, 'active_personnel')
df = clean_numeric_column(df, 'reserve_personnel')
df = clean_numeric_column(df, 'paramilitary')
df = clean_numeric_column(df, 'total_military_aircraft')
df['total_military_aircraft']

df = clean_numeric_column(df, 'fighter_aircraft')
df['fighter_aircraft']

df = clean_numeric_column(df, 'attack_aircraft')
df['attack_aircraft']

df = clean_numeric_column(df, 'transport_aircraft')
df['transport_aircraft']

df = clean_numeric_column(df, 'trainer_aircraft')
df['trainer_aircraft']

df['special_mission_aircraft']
df['tanker_aircraft']


df = clean_numeric_column(df, 'attack_helicopters')
df['attack_helicopters']

df = clean_numeric_column(df, 'tanks')
df['tanks']

df = clean_numeric_column(df, 'armored_fighting_vehicles')
df['armored_fighting_vehicles']

df = clean_numeric_column(df, 'self_propelled_artillery')
df['self_propelled_artillery']

df = clean_numeric_column(df, 'towed_artillery')
df['towed_artillery']

df = clean_numeric_column(df, 'rocket_projectors')
df['rocket_projectors']

df = clean_numeric_column(df, 'total_naval_fleet')
df['total_naval_fleet']

df = clean_numeric_column(df, 'total_naval_fleet_tonnage_mt')
df['total_naval_fleet_tonnage_mt']

df[[
    'aircraft_carriers',
    'helicopter_carriers',
    'submarines',
    'destroyers',
    'frigates',
    'corvettes',
    'coastal_patrol_craft',
    'mine_warfare_craft'
]].dtypes


df = clean_text_column(df, 'defense_budget_usd')
df['defense_budget_usd']

df = clean_text_column(df, 'external_debt_usd')
df['external_debt_usd']

df = clean_text_column(df, 'purchasing_power_parity_usd')
df['purchasing_power_parity_usd']

df = clean_text_column(df, 'foreign_exchange_and_gold_reserves_usd')
df['foreign_exchange_and_gold_reserves_usd']

df = clean_numeric_column(df, 'total_serviceable_airports')
df['total_serviceable_airports']

df = clean_numeric_column(df, 'labour_force')
df['labour_force']

df = clean_numeric_column(df, 'major_ports_and_terminals')
df['major_ports_and_terminals']

df = clean_numeric_column(df, 'total_merchant_marine_fleet')
df['total_merchant_marine_fleet']

df = clean_numeric_column(df, 'railway_coverage_km')
df['railway_coverage_km']

df = clean_numeric_column(df, 'oil_production_bbl')
df['oil_production_bbl']

df = clean_numeric_column(df, 'oil_consumption_bbl')
df['oil_consumption_bbl']

df = clean_numeric_column(df, 'proven_oil_reserves_bbl')
df['proven_oil_reserves_bbl']

df = clean_numeric_column(df, 'natural_gas_production_cum')
df['natural_gas_production_cum']

df = clean_numeric_column(df, 'natural_gas_consumption_cum')
df['natural_gas_consumption_cum']

df = clean_numeric_column(df, 'proven_natural_gas_reserves_cum')
df['proven_natural_gas_reserves_cum']

df = clean_numeric_column(df, 'coal_production_cum')
df['coal_production_cum']

df = clean_numeric_column(df, 'coal_consumption_mt')
df['coal_consumption_mt']

df = clean_numeric_column(df, 'proven_coal_reserves_cum')
df['proven_coal_reserves_cum']

df = clean_numeric_column(df, 'total_land_area_sq_km')
df['total_land_area_sq_km']

df = clean_numeric_column(df, 'coastline_coverage_km')
df['coastline_coverage_km']

df = clean_numeric_column(df, 'border_coverage_km')
df['border_coverage_km']

df = clean_numeric_column(df, 'waterway_coverage_km')
df['waterway_coverage_km']
```
