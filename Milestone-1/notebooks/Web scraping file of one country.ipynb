# This script scrapes military strength data for India from Global Firepower and saves it into a CSV file.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://www.globalfirepower.com/country-military-strength-detail.php?country_id=india"

def scrape_india():
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(URL, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    data = {"country": "India"}

    # Rank
    rank = soup.find("div", class_="country_rank")
    if rank:
        data["power_index_rank"] = rank.text.strip()

    # Helper function
    def get_value(label):
        tag = soup.find(string=lambda x: x and label in x)
        return tag.find_next().text.strip() if tag else None

    data["total_manpower"] = get_value("Total Population")
    data["active_personnel"] = get_value("Active Personnel")
    data["reserve_personnel"] = get_value("Reserve Personnel")
    data["total_aircraft"] = get_value("Aircraft Strength")
    data["fighter_aircraft"] = get_value("Fighters")
    data["naval_assets"] = get_value("Naval Assets")
    data["tanks"] = get_value("Tanks")
    data["defense_budget"] = get_value("Defense Budget")
    data["gdp"] = get_value("Purchasing Power")

    return data


# Run scraper
data = scrape_india()

df = pd.DataFrame([data])
df.to_csv("output.csv", index=False)

print("CSV generated successfully!")
print(df)
```
