### Step 1: Import Required Libraries

We begin by importing the libraries required for web scraping and data processing.

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

### Step 2: Load Country URLs

The list of country pages is stored in the file:

links_for_military_data.txt

Each URL corresponds to a page containing military capability metrics for a specific country.

```python
url = "https://www.globalfirepower.com/country-military-strength-detail.php?country_id=india"
```

### Step 3: Define Metric Extraction code

This extracts military metrics from the HTML page.

The scraper parses metric blocks and stores them into a dictionary structure.

```python
response = requests.get(url, timeout=15)
response.raise_for_status()

soup = BeautifulSoup(response.text, "lxml")

data = {}

title = soup.find("h1")
data["country"] = title.get_text(strip=True)


overview_text = soup.find("span", class_="textNormal")

if overview_text:
    text = overview_text.get_text(" ", strip=True)
    rank_match = re.search(r"ranked\s+(\d+)", text)
    score_match = re.search(r"score of\s+([\d.]+)", text)

    data["power_index_rank"] = rank_match.group(1) if rank_match else None
    data["power_index_score"] = score_match.group(1) if score_match else None

# -----------------------------
# METRIC EXTRACTION
# -----------------------------
metric_blocks = soup.find_all("div", class_="specsGenContainers")

for block in metric_blocks:
    label = block.find("span", class_="textLarge")
    value = block.find("span", class_="textWhite")

    if label and value:
        key = (
            label.get_text(strip=True)
            .replace(":", "")
            .lower()
            .replace(" ", "_")
        )
        data[key] = value.get_text(strip=True)
```

### Step 4: Convert Scraped Data into DataFrame

The collected dictionary objects are converted into a structured pandas DataFrame.

```python
# -----------------------------
# PRINT OUTPUT
# -----------------------------
print("India Data Extracted:\n")
for k, v in data.items():
    print(f"{k}: {v}")


# Save CSV
df = pd.DataFrame([data])     
df.to_csv("data/india_test.csv", index=False)
print("Saved to data/india_test.csv")
```

India’s military data was successfully scraped, structured, and saved as a CSV file.

This step reads all country CSV files from the data folder and combines them into a single dataset.

```python
folder_path = "data"

all_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

df_list = []

for file in all_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)

combined_df
```

All individual country datasets were successfully merged into one combined dataset.

This step exports the combined dataset into a single CSV file for further cleaning and analysis.

```python
combined_df.to_csv("Military_raw_data.csv", index=False)
print("Combined raw dataset saved as Military_raw_data.csv")
```
