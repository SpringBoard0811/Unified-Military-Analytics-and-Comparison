```python
import pandas as pd
import numpy as np
import re

# ==============================
# 1️⃣ Load Dataset
# ==============================
df = pd.read_csv("military_raw_collected (1).csv")  # Upload file in Colab first
df = pd.read_csv(file_path)

print("Original Shape:", df.shape)

# ==============================
# 2️⃣ Standardize Column Names
# ==============================
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# ==============================
# 3️⃣ Remove Duplicates (Country Level)
# ==============================
if "country" in df.columns:
    df = df.drop_duplicates(subset=["country"])

print("After removing duplicates:", df.shape)

# ==============================
# 4️⃣ Clean Numeric Columns
# ==============================

def clean_numeric(value):
    if pd.isna(value):
        return np.nan

    # Convert to string
    value = str(value)

    # Remove commas, %, +, $, and text
    value = re.sub(r"[,+%$]", "", value)

    # Remove words like billion, million, etc.
    value = re.sub(r"(billion|million|trillion)", "", value, flags=re.IGNORECASE)

    # Keep only digits and decimal
    value = re.sub(r"[^0-9.]", "", value)

    return value

# Apply cleaning to all columns except 'country'
for col in df.columns:
    if col != "country":
        df[col] = df[col].apply(clean_numeric)

# ==============================
# 5️⃣ Convert to Numeric Types
# ==============================
for col in df.columns:
    if col != "country":
        df[col] = pd.to_numeric(df[col], errors="coerce")

# ==============================
# 6️⃣ Check Null Percentage
# ==============================
null_percent = df.isnull().mean() * 100
print("\nNull Percentage per Column:")
print(null_percent)

# Keep only columns with < 2% null (optional strict filter)
# df = df.loc[:, null_percent < 2]

# ==============================
# 7️⃣ Validate Row Count
# ==============================
if "country" in df.columns:
    unique_countries = df["country"].nunique()
    print("\nUnique Countries:", unique_countries)

# ==============================
# 8️⃣ Final Info
# ==============================
print("\nFinal Data Types:")
print(df.dtypes)

print("\nFinal Shape:", df.shape)

# ==============================
# 9️⃣ Export Clean CSV
# ==============================
clean_file = "military_cleaned_powerbi.csv"
df.to_csv(clean_file, index=False)

print(f"\nCleaned file saved as {clean_file}")
```
